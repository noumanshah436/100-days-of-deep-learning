```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Conv2D, Flatten, AveragePooling2D
from keras.models import Sequential
from keras.datasets import mnist

# Load dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print("MNIST data loaded successfully!")
```

```python
# LeNet-5 Architecture
model = Sequential()  # Initialize a Sequential model (stack of layers)

model.add(Conv2D(6, kernel_size=(5,5), padding='valid', activation='tanh', input_shape=(32,32,1)))  # First convolution layer with 6 filters of size 5x5
model.add(AveragePooling2D(pool_size=(2,2), strides=2, padding='valid'))  # First average pooling layer to reduce spatial size

model.add(Conv2D(16, kernel_size=(5,5), padding='valid', activation='tanh'))  # Second convolution layer with 16 filters of size 5x5
model.add(AveragePooling2D(pool_size=(2,2), strides=2, padding='valid'))  # Second average pooling layer to reduce size further

model.add(Flatten())  # Flatten 2D feature maps into a 1D vector for dense layers
model.add(Dense(120, activation='tanh'))  # Fully connected layer with 120 neurons
model.add(Dense(84, activation='tanh'))   # Fully connected layer with 84 neurons
model.add(Dense(10, activation='softmax'))  # Output layer with 10 neurons (for 10 digit classes)
```
